### Plotting Lesson

```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
distance_file = os.path.join('distance_data_headers.csv')
distances = np.genfromtxt(fname = distance_file, delimiter = ',', dtype = 'unicode')
print (distances)
```

```python
headers = distances[0]
data = distances[1:]
data = data.astype(np.float)
print (headers)
print (data)
```

```python
plt.figure()
plt.plot(data[:,0],data[:,1])
```

```python
sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (A)')
fig_1 = plt.plot(data[:,0],data[:,1], label = sample)
plt.legend()
plt.savefig('%s.png' % sample)
```

```python
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Å)')
plt.plot(data[:,0],data[:,1], label = headers[1])
plt.plot(data[:,0],data[:,2], label = headers[2])
plt.legend()
#plt.savefig('%s.png' % sample)
```

```python
plt.figure()
for col in range(1,len(headers)):
    fig = plt.plot(data[:,0],data[:,col],label = headers[col])
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (Å)')
plt.legend()
#plt.savefig('%s.png' % sample)
```

```python
for col in range(1,len(headers)):
    plt.figure()
    fig = plt.plot(data[:,0],data[:,col],label = headers[col])
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (Å)')
    plt.legend()
    plt.savefig('%s.png' % headers[col], dpi = 300)
```
